تنظيف البيانات لكي تسهل قرائتها 

```python
KSA['Event name'].str.split('(').str.get(0)
```

```python
KSA["Event name"] = KSA['Event name'].str.split('(').str.get(0)
```

```python
KSA.head()
```

```python
# حذف بعض الاعمده غير المهمه
KSA = KSA.drop(['Athlete age category', 'Year of event', 'Athlete club','Athlete year of birth'], axis = 1)
```

```python
KSA.head()
```

```python
# البحث عن قيم مفقوده
KSA.isna().sum()
```

```python
# بحث عن قيم متكرره
KSA[KSA.duplicated() == True ]
```

```python
# اعاده ترتيب البيانات 
KSA.reset_index(drop = True)
```

```python
KSA.dtypes
```

```python
# تعديل لنوع البيانات 
KSA['Athlete average speed']= KSA['Athlete average speed'].astype(float)
```
